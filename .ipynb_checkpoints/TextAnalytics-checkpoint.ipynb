```python
text_analytics_base_url = "https://southeastasia.api.cognitive.microsoft.com/text/analytics/v2.0/"
```

## Detect language

```python
language_api_url = text_analytics_base_url + "languages"
print(language_api_url)
```

```python
documents = { 'documents': [
    { 'id': '1', 'text': 'This is a document written in English.' },
    { 'id': '2', 'text': 'Este es un document escrito en Español.' },
    { 'id': '3', 'text': '这是一个用中文写的文件' },
    { 'id': '4', 'text': 'Ce texte est en français.' }
]}
print(documents)
```

```python
# The pprint module provides a capability to “pretty-print” arbitrary Python data structures in a form which can be used as input to the interpreter. 
# requests is for HTTP requests and response
import requests
from pprint import pprint
headers   = {"Ocp-Apim-Subscription-Key": subscription_key}
response  = requests.post(language_api_url, headers=headers, json=documents)
languages = response.json()
pprint(languages)
```

```python
#The Python join() method is a string method, and takes a list of things to join with the string. A simpler example might help explain:
from IPython.display import HTML
table = []
for document in languages["documents"]:
    text  = next(filter(lambda d: d["id"] == document["id"], documents["documents"]))["text"]
    langs = ", ".join(["{0}({1})".format(lang["name"], lang["score"]) for lang in document["detectedLanguages"]])
    table.append("<tr><td>{0}</td><td>{1}</td>".format(text, langs))
HTML("<table><tr><th>Text</th><th>Detected languages(scores)</th></tr>{0}</table>".format("\n".join(table)))
```

## Analyze Sentiment

```python
sentiment_api_url = text_analytics_base_url + "sentiment"
print(sentiment_api_url)
```

```python
documents = {'documents' : [
  {'id': '1', 'language': 'en', 'text': 'I had a wonderful experience! The rooms were wonderful and the staff was helpful.'},
  {'id': '2', 'language': 'en', 'text': 'I had a terrible time at the hotel. The staff was rude and the food was awful.'},  
  {'id': '3', 'language': 'es', 'text': 'Los caminos que llevan hasta Monte Rainier son espectaculares y hermosos.'},  
  {'id': '4', 'language': 'es', 'text': 'La carretera estaba atascada. Había mucho tráfico el día de ayer.'}
]}
```

```python
headers   = {"Ocp-Apim-Subscription-Key": subscription_key}
response  = requests.post(sentiment_api_url, headers=headers, json=documents)
sentiments = response.json()
pprint(sentiments)
```

```python
table = []
for document in sentiments["documents"]:
    text  = next(filter(lambda d: d["id"] == document["id"], documents["documents"]))["text"]
    sents = round(document["score"],3)
    table.append("<tr><td>{0}</td><td>{1}</td>".format(text, sents))
HTML("<table><tr><th>Text</th><th>Sentiment (scores)</th></tr>{0}</table>".format("\n".join(table)))
```

## Text Analytics with the resumes

```python
def para2text(p):
    rs = p._element.xpath('.//w:t')
    return u" ".join([r.text for r in rs])
```

```python
import os
import docx
import zipfile
title = os.listdir('Resumefromblob')
text = [para2text(docx.Document('Resumefromblob/' + filename)) for filename in title]
idb=[]
for i in range(len(text)):
    idb.append(i+1)
print(idb)
```

```python
import json
title_text=[{"id":a,"title": b, "text": c} for a,b,c in zip(idb, title, text)]
js= json.dumps(title_text)
doc={}
doc['documents']= title_text
#json string containing id, title of the CV and the content (text) of the CV
json_doc = json.dumps(doc) 

text_id=[{"id":a,"text": b} for a,b in zip(idb, text)]
docbis={}
docbis['documents']= text_id
#json string containing id, title of the CV and the content (text) of the CV
json_docbis = json.dumps(docbis) 
#convert in dictionary
docs=json.loads(json_docbis)

documents=json.loads(json_doc)
print(documents)
```

```python
import pandas as pd
idb,title,text=[],[],[]
for doc in documents['documents']:
    idb.append(doc['id'])
    title.append(doc['title'])
    text.append(doc['text'])
df=pd.DataFrame([title,text], columns=idb, index=['Title','Text']).T
df
```
